Creating Visualizations using Pandas Library 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\PandasYouTubeSeries-main\PandasYouTubeSeries-main\Ice Cream Ratings.csv")
df=df.set_index("Date")
df
```

```python
df.plot(kind="line")
```

```python
df.plot(kind="line",subplots=True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind="bar")
```

```python
df.plot.barh()
```

A stacked bar chart is a type of bar chart that portrays the compositions and comparisons of several variables through time. Stacked charts usually represent a series of bars or columns stacked on top of one another. They are widely used to effectively portray comparisons of total values across several categories.

```python
df.plot(kind="bar",stacked=True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Yellow')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
